```python
plt.figure(figsize = (10,8))
sns.catplot(x = 'quality',data = Dataset , kind = 'count')
```

## __THE GRAPH  ABOVE ILLUSTRATES THE QUALITY VALUES OF THE WINE .  THESE QUALITY   VALUES ARE SIX AND THEY ARE 3,4,5,6,7 & 8 . IF THE WINE QUALITY IS LESS THAN 3, THE WINE IS  BAD AND IF GREATER THEN 8 THE WINE IS CONSIDERED GOOD__

```python
 plt.figure(figsize = (10,8))
sns.barplot(x = 'quality', y ='volatile acidity', data = Dataset , color = 'orange' )
```

## __CONSIDERING THE GRAPH ABOVE, THE QUALITY OF THE WINE INCREASES WITH DECREASING VOLATILE ACIDITY__

```python
plt.figure(figsize = (10,8))
sns.barplot(x = 'quality', y ='density', data = Dataset , color = 'red' )
```

## __SINCE THE BAR DIAGRAM ABOVE IS  AN ISOBAR(HAVING THE SAME HEIGHTS), THERE IS A HIGHER FREQUENCY  OF THE   A PARTICULAR  DENSITY AND THEREFORE THE DATA IS DISTRIBUTED  ACCROSS THE RANGE__ 

```python
corr= Dataset.corr()
plt.figure(figsize=(20,10))
sns.heatmap(corr,annot=True,cmap='coolwarm')
```

# __THE DIAGRAM ABOVE  DEPICTS THE CORRELATION  EXISTING BETWEEN THE VARIOUS COLUMN VALUES IN THE DATASET__ 

```python
plt.plot(Dataset.pH , Dataset.quality)
plt.title('A PLOT OF pH AGAINST QUALITY')    
plt.xlabel( 'pH' )
plt.ylabel('QUALITY')
plt.show()         
```
